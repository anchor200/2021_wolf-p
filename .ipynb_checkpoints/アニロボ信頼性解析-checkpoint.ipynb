```r
nume <- function(data){
    for (i in 1:length(data[1,])){
        data[i,] = as.numeric(data[i,])
    }
    print(data)
    return(data)
    
}


AB <- nume(data_full[str_detect(colnames, pattern="AB_")])
ANT <- nume(data_full[str_detect(colnames, pattern="ANT_")])
REQ <- nume(data_full[str_detect(colnames, pattern="REQ_H")])
RVA <- nume(data_full[str_detect(colnames, pattern="RVA")])
UNK <- nume(data_full[str_detect(colnames, pattern=".UNK")])
THR <- nume(data_full[str_detect(colnames, pattern=".THR")])
UNC <- nume(data_full[str_detect(colnames, pattern=".UNC")])
ETH <- nume(data_full[str_detect(colnames, pattern=".ETH")])
BEN <- nume(data_full[str_detect(colnames, pattern=".BEN")])
SIM <- nume(data_full[str_detect(colnames, pattern=".SIM")])
SBS <- nume(data_full[str_detect(colnames, pattern="SBS")])
IRI <- nume(data_full[str_detect(colnames, pattern=".RI")])
#MEX <- nume(data_full[str_detect(colnames, pattern="MEX")])
RE <- nume(data_full[str_detect(colnames, pattern="RE")])


rev_check <- function(data_mat, min, max){
    soe <- str_detect(colnames(data_mat), pattern="rev")
    soe <- t(matrix(rep(soe, length(data_mat[,1])),ncol=length(data_mat[,1])))
    data_mat_r <- (1 - soe) * data_mat + soe * (max+1 - data_mat)
    return(data_mat_r)
}

ANT <- rev_check(ANT, 1, 7)
UNK <- rev_check(UNK, 1, 7)
THR <- rev_check(THR, 1, 7)
ANT
```
